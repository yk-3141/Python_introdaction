**セットの特徴**

・重複した値を格納できない

・添字やキーの概念がなく、特定の要素に対して代入・参照する方法が存在しない

・添字がないため、要素は順序を持たない

・append関数ではなくadd関数で要素を追加する


```python
# 2-22 セットの利用
scores = {70, 80, 55, 80}
scores.add(80)
print(scores)
print(f'要素数は{len(scores)}')
print(f'合計は{sum(scores)}')
```

```python
# 2-23 コンテナの相互変換
scores = {'network':60, 'database':80, 'security':60}
members = ['松田', '浅木', '工藤']
print(tuple(members)) # リストmembersをタプルに変換して表示
print(list(scores)) # scoresのキーをリストに変換して表示
print(set(scores.values())) # scoresの値をセットに変換して表示
```

```python
# 2-26 2次元リストの例
a = [1, 2, 3]
b = [4, 5, 6]
c = [a, b] # aを0番目、bを1番目とする2次元リスト

print(c) # リストc全体を参照
print(c[0]) # リストcの0番目（リストa）だけを参照
print(c[1][2]) # リストcの1番目（リストb）の2番目だけを参照
```

```python
# 2-28 4つの集合演算
A = {1, 2, 3, 4}
B = {2, 3, 4, 5}
print(A | B)
print(A & B)
print(A - B)
print(A ^ B)
```

**第2章まとめ**

* 複数の値をまとめて扱うコンテナという仕組みがある

* 代表的なコンテナには4つの種類があり、データの特性や管理目的に合わせて適切なコンテナを用いる必要がある

* コンテナには数値や文字列を混在して格納できる

* コンテナには、変数と同じく決まったデータ型を持たない

* リストやタプルで用いる添字は0から始まる
